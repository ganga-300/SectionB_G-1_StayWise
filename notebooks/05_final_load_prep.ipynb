# Hotel Bookings Data: Final Processing and Export

This notebook processes the cleaned hotel bookings dataset to create final, structured outputs ready for business intelligence tools, dashboards, and downstream systems.

```python
# Import additional necessary libraries for EDA and visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from IPython.display import display
```

```python
import pandas as pd
import os

# Ensure the output directory exists
os.makedirs('/content/outputs', exist_ok=True)

# 1. Load Final Processed Data
# Using the provided path based on available files
df = pd.read_csv('/content/eda_final_dataset.csv')

# 2. Final Data Sanity Checks
print("\n--- Data Sanity Checks ---")
print("Missing values per column:")
print(df.isnull().sum())
print("\nNumber of duplicated rows:", df.duplicated().sum())
print("\nShape of the dataset:", df.shape)
```

```python
# 3. Final Feature Selection
final_df = df[
    [
        #  Identifiers / Context
        'hotel',
        'arrival_date_year',
        'arrival_date_month',
        'country_name',

        #  Customer Info
        'customer_type',
        'customer_segment',
        'is_repeated_guest',
        'total_guests',
        'has_children',
        'total_of_special_requests',

        # Booking Info
        'lead_time',
        'booking_type',
        'is_direct_booking',
        'market_segment',
        'distribution_channel',
        'deposit_type',

        #  Stay Info
        'total_nights',
        'stay_duration_category',
        'is_weekend_stay',

        #  Pricing & Revenue
        'average_daily_rate',
        'price_tier',
        'total_revenue',
        'revenue_per_guest',
        'average_night_value',
        'clv_proxy',

        #  Outcomes
        'is_canceled',
        'reservation_status',

        #  Time Features
        'season',
        'is_peak_season'
    ]
]
print("\n--- Final Feature Selection Complete ---")
print("Shape of final_df:", final_df.shape)
```

```python
# 4. Create Summary Tables
print("\n--- Creating Summary Tables ---")
# Revenue Summary
revenue_summary = df.groupby(
    ['customer_segment', 'season']
).agg(
    total_revenue=('total_revenue', 'sum'),
    avg_revenue=('total_revenue', 'mean'),
    bookings=('hotel', 'count')
).reset_index()

# Cancellation Summary
cancellation_summary = df.groupby(
    ['market_segment']
).agg(
    cancellation_rate=('is_canceled', 'mean'),
    total_bookings=('hotel', 'count')
).reset_index()

# Season Summary
season_summary = df.groupby(
    ['season']
).agg(
    total_revenue=('total_revenue', 'sum'),
    cancellation_rate=('is_canceled', 'mean')
).reset_index()
```

```python
# 5. Create KPI Table (Gold Layer)
print("\n--- Creating KPI Table ---")
kpi_table = {
    "total_revenue": df['total_revenue'].sum(),
    "avg_revenue": df['total_revenue'].mean(),
    "cancellation_rate": df['is_canceled'].mean(),
    "avg_lead_time": df['lead_time'].mean()
}
print("KPI Table Created:")
for k, v in kpi_table.items():
    print(f"  {k}: {v:.2f}")
```

```python
# 6. Export Everything
print("\n--- Exporting Data to CSV ---")
final_df.to_csv('final_dataset.csv', index=False)
revenue_summary.to_csv('revenue_summary.csv', index=False)
cancel_summary.to_csv('cancellation_summary.csv', index=False)
season_summary.to_csv('season_summary.csv', index=False)
```
